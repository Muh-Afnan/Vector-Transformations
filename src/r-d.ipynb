```python
fig, ax = plt.subplots()

# eigenvector 1
ax.quiver(0, 0, 1, 2)

# eigenvector 2
ax.quiver(0, 0, -3, 1)

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.grid()

plt.show()
```
